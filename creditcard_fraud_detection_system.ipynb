```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcard = pd.read_csv("creditcard.csv")
creditcard
```

```python
creditcard.info()
```

```python
creditcard.duplicated().sum()
```

```python
creditcard = creditcard.drop_duplicates()
```

```python
creditcard.duplicated().sum()
```

```python
creditcard.isnull().sum()
```

```python
creditcard["Class"].value_counts()
```

```python
#separating the data

legit = creditcard[creditcard.Class == 0]
fraud = creditcard[creditcard.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditcard.groupby("Class").mean()
```

```python
#Under-Sampling

legit_sample = legit.sample(n=473)
```

```python
creditcard_updated = pd.concat([legit_sample, fraud], axis=0)
creditcard_updated
```

```python
creditcard_updated["Class"].value_counts()
```

```python
creditcard_updated.groupby("Class").mean()
```

```python
#Splitting the dataset

X = creditcard_updated.drop(columns="Class", axis=0)
Y = creditcard_updated["Class"]
```

```python
X
```

```python
Y
```

```python
#Splitting the data into the training data and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#MODEL TRAINING

model = LogisticRegression()
```

```python
#Training the model with training data
model.fit(X_train, Y_train)
```

```python
#MODEL EVALUATION

#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score of training data: ",training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score of test data:", test_data_accuracy)
```
